```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("hapiscore_whr.csv")
df
```

```python
df_long = df.melt(id_vars=['country'], var_name='Year', value_name='Happiness Score')
df_long['Year'] = df_long['Year'].astype(int)
```

```python

sns.set_style("whitegrid")


plt.figure(figsize=(12, 6))
sns.lineplot(data=df_long, x='Happiness Score', y='Year', hue='country', marker='o')


plt.xlabel("Happiness Score (WHR)")
plt.ylabel("Year")
plt.title("Happiness Score Trends (2005-2023)")


plt.legend(title="Country")


plt.show()
```
