```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("income.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.rename(columns=lambda x:x.strip(),inplace=True)
```

```python
print(df.columns)
```

```python

sns.lineplot(x=df["country"], y=df["2022"], marker='o', color='b')

plt.xlabel("country", fontsize=12)
plt.ylabel("2022", fontsize=12)
plt.title("income of 2022", fontsize=14)
plt.grid(True)
plt.show()
```
