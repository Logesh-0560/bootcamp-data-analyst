```python
import re
import pandas as pd


commentary_series = df["Commentary"].head(100)  


bowler_pattern = r"^(\w+) to"
batter_pattern = r"to (\w+)"
shot_type_pattern = r"\b(FOUR|SIX|single|1 run|2 runs|3 runs|no run|OUT)\b"

for commentary in commentary_series:
    if isinstance(commentary, str):  
        bowler = re.search(bowler_pattern, commentary)
        batter = re.search(batter_pattern, commentary)
        shot_type = re.findall(shot_type_pattern, commentary)

        print(f"Commentary: {commentary}")
        print(f"Bowler: {bowler.group(1) if bowler else None}")
        print(f"Batter: {batter.group(1) if batter else None}")
        print(f"Shot Type: {shot_type[0] if shot_type else 'Dot Ball'}\n")

```
